```python
def gencube(n):
  l=[]
  for i in range(n):
    l.append(i**3)
  return l
```

```python
gencube(6)
```

```python
gencube(60)
```

when n is increasing we need lot of memory to store the list l for all the values of cubes.

```python
def gencube(n):
  for i in range(n):
    yield i**3
```

```python
gencube(10)
```

```python
for i in gencube(10):
  print(i)
```

```python
def fibonacci(n):
  a=1
  b=1
  for i in range(n):
    yield a
    a,b=b,a+b
```

```python
for i in fibonacci(10):
  print(i)
```
