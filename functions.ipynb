```python
def tri(r,c):
  """
  this is a function to create a triangle with given rows and column
  """
  for i in range(r):
    for j in range(i+1):
      if j<c:
        print('*',end='')
    print('\r')
```

```python
def dict_list(arg):
    l=[]
    if type(arg)==dict:
      for i in arg.values():
        if type(i)==dict:
          l1=dict_list(i)
          l=l+l1
        else:
          l.append(i)
    return l
```
